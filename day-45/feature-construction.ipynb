```python
Y.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,Y,scoring='accuracy',cv=20))
```

### Applying Feature Construction

```python
X['Family'] = X['SibSp'] + X['Parch'] + 1
```

```python
X.head()
```

```python
def family(num):
    if num == 1:
        #alone
        return 0
    elif num > 1 and num <= 4:
        #small family
        return 1
    else:
        #large family
        return 2
        
        
```

```python
family(8)
```

```python
X['Family_types'] = X['Family'].apply(family)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family']).head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,Y,scoring='accuracy',cv=20))
```
