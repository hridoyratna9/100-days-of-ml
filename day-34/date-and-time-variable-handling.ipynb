### working with Dates

```python
#Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

#### 1.Extract year

```python
date['date_year'] = date['date'].dt.year
date.sample(5)
```

#### 2. Extract month

```python
date['date_month'] = date['date'].dt.month
date.sample(5)
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.sample(5)
```

#### 3. Extract dday

```python
date['date_day'] = date['date'].dt.day
date.sample(5)
```

```python
date['date_day_name'] = date['date'].dt.day_name()
date.sample(5)
```

### 4. day of week

```python
date['date_dow'] = date['date'].dt.dayofweek
date.sample(5)
```

```python
date['date_dow_name'] = date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

#### 5. Extract week of the year

```python
date['date_week'] = date['date'].dt.week
date.drop(columns=['product_id','city_id','orders']).head()
```

### 6. Extract Quarter

```python

date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 7. Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 8. Extract Time elapsed between dates

```python

import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
np.round((today - date['date']) / np.timedelta64(1,'ms'),0)
```

```python
np.round((today - date['date']) / np.timedelta64(1,'us'),0)
```
